```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# Generating some data to plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creating the main figure and axes
fig, ax = plt.subplots()

# Plotting the data on the main axes
ax.plot(x, y, label='Main plot')

# Defining the bounds of the inset (the zoomed area)
x1, x2, y1, y2 = 2, 3, 0, 0.5

# Adding a rectangle to highlight the zoomed area on the main plot
ax.add_patch(plt.Rectangle((x1, y1), x2-x1, y2-y1, fill=False, edgecolor='red', linestyle='--'))

# Creating an inset axes within the main axes
axins = inset_axes(ax, width="30%", height="30%", loc='upper right')

# Plotting the same data on the inset but limiting its range to the zoomed area
axins.plot(x, y)
axins.set_xlim(x1, x2)
axins.set_ylim(y1, y2)

# Optionally, remove the ticks on the inset to make it cleaner
axins.xaxis.set_visible(False)
axins.yaxis.set_visible(False)

# Showing the plot
plt.show()
```
