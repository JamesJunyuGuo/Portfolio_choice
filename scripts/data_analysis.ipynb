```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
M_list = [2000,1000,500,200,100]
N_list =  [100,100,50,20,10]
L_list  = [2000,2000,1000,400,200]

for M,L,N in zip(M_list,L_list,N_list):
    postfix = f'CEVSV;L{L};M{M};N{N}'
    thetau = pd.read_csv(f'./Result/theta_u{postfix}.csv',header=None,index_col=None).to_numpy()[1:M+1][0:M]
    true_thetau = pd.read_csv(f'./Result/true_theta_u{postfix}.csv',header=None,index_col=None).to_numpy()[0:M]
    diff_thetau = pd.read_csv(f'./Result/diff_theta_u{postfix}.csv',header=None,index_col=None).to_numpy()[0:M]
    
    
    
```

```python
for i in range(len(M_list)):
    M = M_list[i]
    N = N_list[i]
    L = L_list[i]
    postfix = f'CEVSV;L{L};M{M};N{N}'
    thetau = pd.read_csv(f'./Result/theta_u{postfix}.csv',header=None,index_col=None).to_numpy()[1:M+1][0:M]
    true_thetau = pd.read_csv(f'./Result/true_theta_u{postfix}.csv',header=None,index_col=None).to_numpy()[0:M]
    diff_thetau = pd.read_csv(f'./Result/diff_theta_u{postfix}.csv',header=None,index_col=None).to_numpy()[0:M]

    absolute_error = np.abs(diff_thetau)[:,0]
    comparative_error = np.abs(diff_thetau/true_thetau)[:,0]

    fig, axs = plt.subplots(2, 1, figsize=(8, 6), dpi=200)
    axs[0].plot(np.linspace(0,1,M),absolute_error,label='Absolute Error')

    # Plot on the second subplot
    axs[1].plot(np.linspace(0,1,M),comparative_error,label='Comparative Error')

    # Add titles to subplots
    axs[0].set_title('Abosolute Error')
    axs[1].set_title('Comparative Error')
    fig.savefig('./Result/theta_u{postfix}.png')
    
```
